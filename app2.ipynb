## Punctuation Removal

Pengilangan simbol khusus seperti \r, \n, ", url, dan byte code

```python
def cleaning(df):    
    df = re.sub(r"\n", " ",df)
    df = re.sub(r"\r", " ",df)
    df = re.sub(r"&amp;", " ",df)
    df = re.sub(r"&gt;", " ",df)
    df = re.sub(r"&lt;", " ",df)
    df = re.sub(r"&quot;", " ",df)
    df = re.sub(r"\u00a0", " ",df) 
    df = re.sub(r'''(?i)\b((?:https|http?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', "",df)
    df = re.sub(r"[-_<>*#$@()\"/%&;:<>{}`+=~|.!?,']", " ", df) #hapus simbol
    df = re.sub(r"[^a-zA-Z]", " ",df) # menghilangkan angka
    df = re.sub("\s\s+", " ",df) # menghilangkan double spasi
    return df
```

## Word Normalization

```python
kamus_normal = pd.read_excel("./static/data/excel/normalisasi.xlsx",header=None,names=["non-standard word","standard word"])
kamus_normal
```

```python
nonstdword = kamus_normal['non-standard word'].values.tolist()
stdword = kamus_normal['standard word'].values.tolist()
```

```python
def normalize_text(text, stdword_, nonstdword_):
    text = text.split(" ")
    for i in range(len(text)):
        if text[i] in nonstdword_:
            index = nonstdword_.index(text[i])
            text[i] = stdword_[index]
    return ' '.join(map(str, text))

def normalizeData(data, stdword_, nonstdword_):
    data = data.map(lambda com : normalize_text(com,stdword_,nonstdword_))
    return data
```

## Meaningless Removal

```python
meaningless = pd.read_excel("./static/data/excel/meaningless.xlsx",header=None,names=['stopword'])
meaningless
```

```python
def removeMeaningless(data):
    meaningless = pd.read_excel("./static/data/excel/meaningless.xlsx",header=None,names=['stopword'])
    meaningless = meaningless['stopword'].tolist()
    for word in meaningless:
        data = re.sub(r"\b" + word + r"\b" , " ",data)
    return data
```

## Stopword Removal (Sastrawi)

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
stop_words = StopWordRemoverFactory().get_stop_words() # kata-kata stopword sastrawi indonesia
```

```python
def removeStopword(df,stop_words_):
    for stop_word in stop_words_:
        regex_stopword = r"\b" + stop_word + r"\b"
        df = df.str.replace(regex_stopword, '')
        # df = re.sub("\s+", " ",df)
    return df
```

## Stemming (Sastrawi)

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
stemmerID = StemmerFactory().create_stemmer() # sastrawi
```

```python
def stemming(text, stemmer_id):
    text_split = text.split(" ")
    stemmed_list = []
    for i in text_split:
        stemmed_list.append(stemmer_id.stem(i))
    stemmed = ' '.join(map(str,stemmed_list))
    return stemmed

def stemData(data, stemmer_id):
    data = data.map(lambda com : stemming(com,stemmer_id))
    # data = re.sub("\s\s+", " ",data)
    return data
```

## Preprocessing

```python
import re
def preprocessing(df):
    df['cleaning'] = df['content'].apply(cleaning)
    df['case_folding'] = df['cleaning'].apply((str.lower))
    df['normalize'] = normalizeData(df['case_folding'], stdword, nonstdword)
    df['meaningless_removal'] = df['normalize'].apply(removeMeaningless)
    df['remove_stopword'] = removeStopword(df['meaningless_removal'], stop_words).apply(cleaning)  # no stemming, stopword
    df['stemming'] = stemData(df['remove_stopword'], stemmerID)
    df['after_preprocessing'] = df['stemming']
    return df

df = preprocessing(df)
```

```python
service = ['mudah', 'cepat', 'lama', 'efisien', 'bagus', 'mantap', 'bantu', 'keren', 'baik', 'error', 'waktu', 'aman', 'lancar', 'guna', 'nyaman']
doctor = ['ramah', 'konsul', 'responsif', 'diagnogsa', 'komunikatif', 'respon', 'tepat', 'chat', 'spesialis', 'akurat', 'keluhan', 'pasien', 'temu', 'tanya']
medicine = ['resep', 'beli', 'sakit', 'pesan', 'cocok', 'apotek', 'klinik', 'sembuh', 'antar', 'kirim', 'butuh', 'sesuai']
fee = ['mahal', 'murah', 'terjangkau', 'transaksi', 'hemat', 'refund', 'diskon', 'minim', 'mampu', 'besar', 'bayar', 'promo']

def labelling(data):
    data.reset_index(inplace=True)
    service_=[]
    doctor_ = []
    medicine_ = []
    fee_ = []

    for i in range(len(data['after_preprocessing'])):
        if any(e_service in data.get('after_preprocessing').get(i) for e_service in service):
            if data['score'][i] < 3:
                service_.append(-1)
            elif data['score'][i] > 3:
                service_.append(1)
            else:
                service_.append(0)
        else:
            service_.append(0)

        if any(e_doctor in data.get('after_preprocessing').get(i) for e_doctor in doctor):
            if data['score'][i] < 3:
                doctor_.append(-1)
            elif data['score'][i] > 3:
                doctor_.append(1)
            else:
                doctor_.append(0)
        else:
            doctor_.append(0)

        if any(e_medicine in data.get('after_preprocessing').get(i) for e_medicine in medicine):
            if data['score'][i] < 3:
                medicine_.append(-1)
            elif data['score'][i] > 3:
                medicine_.append(1)
            else:
                medicine_.append(0)
        else:
            medicine_.append(0)

        if any(e_fee in data.get('after_preprocessing').get(i) for e_fee in fee):
            if data['score'][i] < 3:
                fee_.append(-1)
            elif data['score'][i] > 3:
                fee_.append(1)
            else:
                fee_.append(0)
        else:
            fee_.append(0)

    if "service" in data.columns:
        data = data.drop(columns=["service"])
    data.insert(0, "service", service_)
    if "doctor" in data.columns:
        data = data.drop(columns=["doctor"])
    data.insert(0, "doctor", doctor_)
    if "medicine" in data.columns:
        data = data.drop(columns=["medicine"])
    data.insert(0, "medicine", medicine_)
    if "fee" in data.columns:
        data = data.drop(columns=["fee"])
    data.insert(0, "fee", fee_)
    data = data.dropna()
    data = data.drop(data[(data['service'] == 0) & (data['doctor'] == 0) & (data['medicine'] == 0) & (data['fee'] == 0)].index)
    return data


df = labelling(df)
df.head()
```

## TF-RF

```python
import numpy as np

def tfrf(data_train, data_test):
    data_train.dropna(inplace=True)
    data_test.dropna(inplace=True)
    X_train, X_test, y_train, y_test = data_train['after_preprocessing'] , data_test['after_preprocessing'], data_train[['service','doctor','medicine','fee']] , data_test[['service','doctor','medicine','fee']]
    
    
    len_X_train = len(X_train)
    len_X_test = len(X_test)
    sentences = data_train['after_preprocessing'].append(data_test['after_preprocessing'])    
    # Initialize dictionary to store the overall token counts
    overall_token_counts = {}

    # Loop through each sentence
    for sentence in sentences:
        # Tokenize the sentence
        tokens = sentence.split()

        # Count the number of occurrences of each token
        token_counts = {token: tokens.count(token) for token in set(tokens)}

        # Add the token counts to the overall counts
        for token, count in token_counts.items():
            if token in overall_token_counts:
                overall_token_counts[token] += count
            else:
                overall_token_counts[token] = count

    # Create a list of all tokens
    all_tokens = list(overall_token_counts.keys())

    # Create a matrix of zeroes with dimensions of sentences and tokens
    matrix = np.zeros((len(sentences), len(all_tokens)))

    # Loop through each sentence
    for i, sentence in enumerate(sentences):
        # Tokenize the sentence
        tokens = sentence.split()
        # Calculate the relative frequency of each token
        total_tokens = len(tokens)
        token_rf = {token: tokens.count(token)/total_tokens for token in set(tokens)}
        # Add the relative frequency to the matrix
        for j, token in enumerate(all_tokens):
            if token in token_rf:
                matrix[i, j] = token_rf[token]
                
    X_train = matrix[:len_X_train]
    X_test = matrix[len_X_train:]
    
    return X_train, X_test, y_train, y_test
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def tfidf(data_train, data_test):
    data_train.dropna(inplace=True)
    data_test.dropna(inplace=True)
    X_train, X_test, y_train, y_test = data_train['after_preprocessing'] , data_test['after_preprocessing'], data_train[['service','doctor','medicine','fee']] , data_test[['service','doctor','medicine','fee']]
    vectorizer = TfidfVectorizer()
    X_train = vectorizer.fit_transform(X_train).toarray()
    X_test = vectorizer.transform(X_test).toarray()
    
    return X_train, X_test, y_train, y_test
```

# **Classification**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multioutput import MultiOutputClassifier
def classification(X_train, X_test, y_train, y_test, k):
    clf = MultiOutputClassifier(KNeighborsClassifier(n_neighbors=k))
    clf.fit(X_train, y_train)

    pred = clf.predict(X_train) # predict data train
    y_pred = clf.predict(X_test) # predict data test
    
    return X_train, X_test, y_train, y_test, y_pred, pred, clf
```

```python
import numpy as np
def accuracy(y_test, y_pred):
    miss_pred = np.sum(np.not_equal(np.array(y_test), np.array(y_pred)))/float(y_test.size)
    accuracy = 1-miss_pred
    return round((accuracy * 100),2)
```

```python
from sklearn.metrics import f1_score
def F1Score(y_test, y_pred):
    price = f1_score(y_test['service'], y_pred[:,0], average='macro')
    pack = f1_score(y_test['doctor'], y_pred[:,1], average='macro')
    prod = f1_score(y_test['medicine'], y_pred[:,2], average='macro')
    aroma = f1_score(y_test['fee'], y_pred[:,3], average='macro')
    average = (price + pack + prod + aroma) / 4
    return round((average * 100),2)
```

```python
from sklearn.metrics import precision_score
def precisionScore(y_test, y_pred):
    price = precision_score(y_test['service'], y_pred[:,0], average='macro')
    pack = precision_score(y_test['doctor'], y_pred[:,1], average='macro')
    prod = precision_score(y_test['medicine'], y_pred[:,2], average='macro')
    aroma = precision_score(y_test['fee'], y_pred[:,3], average='macro')
    average = (price + pack + prod + aroma) / 4
    return round((average * 100),2)
```

```python
from sklearn.metrics import recall_score
def recallScore(y_test, y_pred):
    price = recall_score(y_test['service'], y_pred[:,0], average='macro')
    pack = recall_score(y_test['doctor'], y_pred[:,1], average='macro')
    prod = recall_score(y_test['medicine'], y_pred[:,2], average='macro')
    aroma = recall_score(y_test['fee'], y_pred[:,3], average='macro')
    average = (price + pack + prod + aroma) / 4
    return round((average * 100),2)
```

```python
def findBestK(X_train, X_test, y_train, y_test):
    bestK=1
    bestAccuracy=0
    for i in range(1,20,2):
        X_train_, X_test_, y_train_, y_test_, y_pred, pred, clf = classification(X_train, X_test, y_train, y_test, i)
        if accuracy(y_test_, y_pred) > bestAccuracy:
            bestAccuracy = accuracy(y_test_, y_pred)
            bestK = i
    return bestK
```
